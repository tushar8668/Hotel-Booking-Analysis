# **EDA-Hotel Booking Analysis**

  Have you ever wondered when the best time of year to book a hotel room is? Or the optimal length of stay in order to get the best daily rate? What if you wanted to predict whether or not a hotel was likely to receive a disproportionately high number of special requests? This hotel booking dataset can help you explore those questions!.




This data set contains booking information for a city hotel and a resort hotel, and includes information such as when the booking was made, length of stay, the number of adults, children, and/or babies, and the number of available parking spaces, among other things. All personally identifying information has been removed from the data.

[Github link](https://github.com/tushar8668/Hotel-Booking-Analysis/tree/main)

# **Dataset Information** 


- hotel: Name of hotel ( City or Resort)
- is_canceled: Whether the booking is canceled or not (0 for no canceled and 1 for canceled)
- lead_time: time (in days) between booking transaction and actual arrival.
- arrival_date_year: Year of arrival
- arrival_date_month: month of arrival
- arrival_date_week_number: week number of arrival date.
- arrival_date_day_of_month: Day of month of arrival date
- stays_in_weekend_nights: No. of weekend nights spent in a hotel
- stays_in_week_nights: No. of weeknights spent in a hotel
- adults: No. of adults in single booking record.
- children: No. of children in single booking record.
- babies: No. of babies in single booking record. 
- meal: Type of meal chosen 
- country: Country of origin of customers (as mentioned by them)
- market_segment: What segment via booking was made and for what purpose.
- distribution_channel: Via which medium booking was made.
- is_repeated_guest: Whether the customer has made any booking before(0 for No and 1 for  Yes)
- previous_cancellations: No. of previous canceled bookings.
- previous_bookings_not_canceled: No. of previous non-canceled bookings.
- reserved_room_type: Room type reserved by a customer.
- assigned_room_type: Room type assigned to the customer.
- booking_changes: No. of booking changes done by customers
- deposit_type: Type of deposit at the time of making a booking (No deposit/ Refundable/ No refund)
- agent: Id of agent for booking
- company: Id of the company making a booking
- days_in_waiting_list: No. of days on waiting list.
- customer_type: Type of customer(Transient, Group, etc.)
- adr: Average Daily rate.
- required_car_parking_spaces: No. of car parking asked in booking
- total_of_special_requests: total no. of special request.
- reservation_status: Whether a customer has checked out or canceled,or not showed 
- reservation_status_date: Date of making reservation status.

### Cleaning the Dataset

Cleaning the dataset is most important part of exploratery data anlysis. That kind of unclean data can affect our outcome of EDA. So we clean the data first.



1.   first we remove duplicated entries from our data




```python
hotel_data[hotel_data.duplicated()].shape
```

```python
hotel_data.drop_duplicates(inplace = True)
```

2.   Remove the null values 



```python
hotel_data.isnull().sum().sort_values(ascending = False)[:5]
```

Here, compony and agent having int data type that means they have compony number and agent number as data in it. There are some cases when customer didnt booked hotel via agent or compony so in that case the value is set to be null under these columns, so we replace null value by 0 in these columns.



```python
hotel_data[['company','agent']] = hotel_data[['company','agent']].fillna(0)
```

```python
hotel_data['children'].unique()
```

We see in column children has 0 value means there are 0 children present, and 'nan' are the missing values.
So we fill this missing values in column by its means. 

```python
hotel_data['children'].fillna(hotel_data['children'].mean(), inplace = True)
```

Now last column is country in our dataset which having null values, so we replace those missing values by 'other'.

```python
hotel_data['country'].fillna('others', inplace = True)
```

```python
hotel_data.isnull().sum().sort_values(ascending = False)[:6]
```

3.   Convert the dataype of columns

 9.  According to the hotels, Which month generates the high revenue ?

```python
r = hotel_data[hotel_data['hotel'] == 'Resort Hotel']
resort_data = r.groupby(['arrival_date_month'])['adr'].mean().reset_index()
c = hotel_data[hotel_data['hotel'] == 'City Hotel']
city_data = c.groupby(['arrival_date_month'])['adr'].mean().reset_index()
hotels_data = resort_data.merge(city_data , on='arrival_date_month')
hotels_data.columns = ['Months','price for resort hotel','price for city hotel']
Months =months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
hotels_data['Months'] = pd.Categorical(hotels_data['Months'] , categories= Months , ordered=True)
hotels_data = hotels_data.sort_values('Months').reset_index().drop('index',axis=1)
hotels_data
```

```python
# Plotting the line chart
plt.figure(figsize=(14,7))
sns.lineplot(x = 'Months', y = 'price for resort hotel' ,data=hotels_data)
sns.lineplot(x = 'Months', y = 'price for city hotel' , data=hotels_data)
plt.ylabel('Average Rate')
plt.legend(['Resrt Hotel','City Hotel'])
plt.show()
```

By the line graph, we observed that resort hotels generated high revenue in 'August' and city hotels are generated high revenue in month of 'May' and 'August'.

13. How Long People Stay in the hotel ?

we create new column in the dataset that shows how many days peaples will stay in hotels.

```python
data = hotel_data.copy()
data['total_stay'] = data['stays_in_week_nights'] + data['stays_in_weekend_nights']
data['total_number_of_peoples'] = data['adults']+data['children'] + data['babies']
data_not_cancel = data[data['is_canceled']==0]
data_not_cancel = data_not_cancel[data_not_cancel['total_stay'] < 14]
plt.figure(figsize = (10,5) )
sns.countplot(x=data_not_cancel['total_stay'] , hue = data['hotel'])
plt.show()
```

17. Not getting same room will affect on adr or not ?

```python
data = hotel_data.copy()


def to_check(x):
  if x['reserved_room_type'] != x['assigned_room_type']:
    return 1
  else:
    return 0

# Create new column as Room alloted or not
data['Not_alloted_same_room'] = data.apply(lambda x : to_check(x),axis = 1)
dataR = data.groupby('is_canceled')
Data_room = pd.DataFrame((dataR['Not_alloted_same_room'].sum()/dataR.size())*100).reset_index().rename(columns = {0 : '%_of_Not_getting_same_room'})


plt.figure(figsize = (10,5))
sns.barplot(x = Data_room['is_canceled'] , y = Data_room['%_of_Not_getting_same_room'])
plt.title('% of cancellation for not getting same room')
plt.show()
```

Not getting same room is not cause to canceled the bookings. There are almost 20% of bookings are not canceled dew to not getting same room.



*   Lets see that not getting same room affect on adr ?

```python
data = data[data['adr'] < 5000]
plt.figure(figsize = (12,6))
sns.boxplot(x = 'Not_alloted_same_room', y = 'adr', data = data)
plt.title('Affect Vs adr Plot')
plt.show()
```

So here we observed that Not alloting the same room is actually affect on adr. Customer who didn't get the same room pay little less, except some cases.
